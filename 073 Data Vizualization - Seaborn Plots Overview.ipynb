# Lineplot

```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

#define the dataframe
fig, ax = plt.subplots(figsize=[10, 5])
df = pd.DataFrame({"product":["sedan", "cabrio", "cabrio", "sportsback", "quatro", "quatro"], "sold":[1,5,3,4,5,10]})

# display the plot
print(df)
sns.lineplot(data = df, y='sold', x=df.index, color='royalblue')
ax.tick_params(axis="x", rotation=0, labelsize=10, length=0)
ax.set_title('Lineplot')
plt.show()
```

# Barplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

# define the dataframe 
df = pd.DataFrame({"fruits":["strawberry", "apple", "peach", "pienaple"], "stock":[1,7,7,4]})

# display the plot
fig, axs = plt.subplots(figsize=[10, 5], ncols=2, nrows=1)
sns.barplot(data=df, y='fruits', x='stock', ax=axs[0])
axs[0].tick_params(axis="x", rotation=0, labelsize=10, length=0)
axs[0].set_title('Pieplot horizontal')
sns.barplot(data=df, x='fruits', y='stock', ax=axs[1])
axs[1].tick_params(axis="x", rotation=0, labelsize=10, length=0)
axs[1].set_title('Pieplot vertical')
plt.show()
```

# Histplot

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# define the dataframe
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# display the plot
print(s)
fig, ax = plt.subplots(figsize=(10, 5))
sns.histplot(s, linewidth=2, color='royalblue', kde=True, ax=ax)
ax.tick_params(axis="x", rotation=0, labelsize=10, length=0)
ax.set_title('Histplot')
plt.show()
```

# Displot (Density Plot)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# define the dataframe
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
z = np.random.normal(mu, sigma, 1000)
f = np.random.normal(mu, sigma, 1000)

df = pd.DataFrame({'s':s, 'z':z, 'f':f})

# display the plot
print(s)

sns.displot(data=df, linewidth=2, kind='kde')
plt.show()
```

# Scatterplot

```python
from matplotlib import pyplot as plt
import random
import seaborn as sns
import pandas as pd

# define the dataframe
x, y, z = [], [], []
for i in range(0,50):
    x.append(random.randint(1,50))
    y.append(random.randint(1,50))
    z.append(random.choice(['a', 'b', 'c']))
df = pd.DataFrame({'x':x, 'y':y, 'z':z})

# display the plot
print(df)
fig, ax = plt.subplots(figsize=[10, 5])
sns.scatterplot(data=df, x=x, y=y, hue=z, color='red')
ax.tick_params(axis="x", rotation=0, labelsize=10, length=0)
ax.set_title('Scatterplot')
plt.show()
```

# Pieplot

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

# define the dataframe 
df = pd.DataFrame({"fruits":["strawberry", "apple", "peach", "pienaple", "blueberry", "cherry"], "sold":[1,5,3,4,5,10]})

# display the plot
colors = sns.color_palette('pastel')[0:6]
fig, ax = plt.subplots(figsize=[7, 7])
plt.pie(data=df, x='sold', labels='fruits', colors=colors)
ax.tick_params(axis="x", rotation=0, labelsize=10, length=0)
ax.set_title('Pieplot')
plt.show()
```

# Heatmap

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd

corr = ([[ 1., -0.04, -0.20, -0.24],
       [-0.05,  1.,  0.12, -0.29],
       [-0.20,  0.4,  1.,  0.25],
       [-0.24, -0.29,  0.28 ,  1.]])

# Generate a custom diverging colormap
cmap = "RdBu"

# Draw the heatmap
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(
    corr,
    cmap=cmap,
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5},
)
```

# Boxplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# define the dataframe
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

df = pd.DataFrame({'s':s})

# box plot design
PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'royalblue'},
    'medianprops':{'color':'coral'},
    'whiskerprops':{'color':'royalblue'},
    'capprops':{'color':'royalblue'}
    }
sns.set_style('ticks', {'axes.edgecolor': 'grey',  
                        'xtick.color': '0',
                        'ytick.color': '0'})

# display the plot
fig, ax = plt.subplots(figsize=[10,2])
sns.boxplot(data=df, orient="h", ax = ax, color='royalblue', flierprops={"marker": "o"}, **PROPS)
plt.show()
```

# Violin Plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# define the dataframe
mu, sigma = 0, 0.1 # mean and standard deviation

f, s, t = [], [], []
for i in range(50):
    f.append(np.random.choice([True, False]))
    t.append(np.random.choice(['A', 'B', 'C']))
    s.append(np.random.randint(1,50))
df = pd.DataFrame({'s':s, 'f':f, 't':t})

# display the plot
fig, ax = plt.subplots(figsize=[10,5])
sns.violinplot(data=df, y='s', x='t', hue='f', split=True, ax=ax)
ax.set_title('Violin plot')
plt.show()
```
